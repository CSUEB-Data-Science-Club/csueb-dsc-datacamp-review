```python
# Use this cell to begin your analysis, and add as many as you would like!
from matplotlib import pyplot as plt
import pandas as pd

plt.rcParams['figure.figsize'] = [11, 7]
```

```python

df = pd.read_csv("datasets/office_episodes.csv")
df.info()
```

```python
fig = plt.figure()
x_col, y_col = 'episode_number', 'viewership_mil'
plt.scatter(df[x_col], df[y_col])
```

```python
colors_by_row = []
for i, row in df.iterrows():
    if row['scaled_ratings'] < 0.25:
        colors_by_row.append('red')
    elif row['scaled_ratings'] >= 0.25 and row['scaled_ratings']  < 0.5:
        colors_by_row.append('orange')
    elif row['scaled_ratings'] >= 0.5 and row['scaled_ratings']  < 0.75:
        colors_by_row.append('lightgreen')
    elif row['scaled_ratings'] >= 0.75:
        colors_by_row.append('darkgreen')


df.head()
```

```python
fig = plt.figure()

plt.scatter(
    x=df[x_col], 
    y=df[y_col],
    c=colors_by_row)
plt.show()
```

```python
print(plt.scatter.__doc__)
```

```python
sizes = [ 250 if x == True else 25 for x in df["has_guests"].to_list()]
print(type(sizes))
sizes[:5]
```

```python
fig = plt.figure()
plt.title("""Popularity, Quality, and Guest Appearances on the Office""")
plt.xlabel("Episode Number")
plt.ylabel("Viewership (Millions)")
plt.grid(True)
plt.scatter(
    x=df[x_col], 
    y=df[y_col],
    c=colors_by_row,
    s=sizes
)
plt.show()
```
